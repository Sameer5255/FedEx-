```python
# Convert 'Scheduled_Delivery_Date' and 'Delivered_to_Client_Date' columns to datetime format.
# Calculate delivery delay in days by subtracting scheduled from actual delivery dates.

data['Scheduled_Delivery_Date'] = pd.to_datetime(data['Scheduled_Delivery_Date'])
data['Delivered_to_Client_Date'] = pd.to_datetime(data['Delivered_to_Client_Date'])

# Create a histogram to visualize the distribution of delivery delays.
# Add a vertical red dashed line at x=0 to indicate on-time deliveries.
# Include KDE for smooth distribution curve, and add grid, labels, and legend for better readability.
data['Delivery_Delay_Days'] = (data['Delivered_to_Client_Date'] - data['Scheduled_Delivery_Date']).dt.days
plt.figure(figsize=(10, 6))
sns.histplot(data['Delivery_Delay_Days'], bins=30, kde=True, color='skyblue')
plt.title("Delivery Delay Distribution")
plt.xlabel("Delay in Days (Delivered - Scheduled)")
plt.ylabel("Number of Orders")
plt.axvline(x=0, color='red', linestyle='--', label='On-time Delivery')
plt.legend()
plt.grid(True)
plt.show()
```

##insight
highest order delivered on scheduled day

```python
# Create an interactive histogram using Plotly to show the count of shipments for each shipment mode.
# Bars are colored by 'Shipment_Mode' for visual distinction.
# The chart title is set to 'Count of Shipments by Mode'.
fig = px.histogram(data, x='Shipment_Mode', color='Shipment_Mode',
                   title='Count of Shipments by Mode',)
fig.show()
```

```python
##Insight
#Mostly Air shipment mode is choosen and least is ocean mode.
```

```python
# Create an interactive box plot using Plotly to show the distribution of delivery delays
# for each shipment mode. Each box represents the spread of 'Delivery_Delay_Days' per mode.
# Colors diff
fig1= px.box(data, x='Shipment_Mode',
                 y='Delivery_Delay_Days',
                 color ='Shipment_Mode',
                   title='Delivery Delay by Shipment Mode',)
fig1.show()
```

##Insight
Air charter mostly delay in delivery if we ignore outliers

```python
# Calculate the average delivery delay for each shipment mode by grouping the data
# by 'Shipment_Mode' and taking the mean of 'Delivery_Delay_Days'.
# The result is reset to a new DataFrame with a clean index.

shipment_delay_avg = data.groupby('Shipment_Mode')['Delivery_Delay_Days'].mean().reset_index()
```

```python
# Create an interactive line chart using Plotly to visualize the average delivery delay
# for each shipment mode. Points (markers) are added to highlight individual mode values.
# The x-axis shows different shipment modes, and the y-axis shows their average delay in days.
fig = px.line(shipment_delay_avg,
              x='Shipment_Mode',
              y='Delivery_Delay_Days',
              title='Average Delivery Delay by Shipment Mode',
              markers=True)  # adds points to the line

fig.show()
```

##Insight
ocean shipment mode mostly delay in delivery if we ignore outliers

```python
# Remove rows with missing values in 'Brand' or 'Unit_Price' columns to clean the data.

data_clean = data.dropna(subset=['Brand', 'Unit_Price'])


# Identify the top 10 most frequent brands based on order counts.


top_brands = data_clean['Brand'].value_counts().nlargest(10).index

# Filter the cleaned data to include only these top brands.
data_top_brands = data_clean[data_clean['Brand'].isin(top_brands)]
```

```python
# Create an interactive box plot with Plotly to visualize the distribution of unit prices for each top brand.
# Colors differentiate brands, and the plot includes a descriptive title.
fig2= px.box(data_top_brands,x='Brand',
             y='Unit_Price',
             color='Brand',
             title='Unit Price Distribution by Brand')
fig2.show()
```

isentress brand have highest unit price  and sustriva has lowest
